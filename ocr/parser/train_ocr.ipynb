```python
import json
import os

BASE_PATH = '/content/drive/MyDrive/OCR'
INPUT_JSON_PATH = os.path.join(BASE_PATH, 'train/label_studio_data.json')
OUTPUT_TXT_PATH = os.path.join(BASE_PATH, 'train/train_txt.txt')

if not os.path.exists(INPUT_JSON_PATH):
    print("오류: JSON 파일을 찾을 수 없습니다:", INPUT_JSON_PATH)
else:
    with open(INPUT_JSON_PATH, 'r', encoding='utf-8') as f:
        data = json.load(f)

    out_lines = []

    for task in data:
        file_upload = task.get('file_upload', '')
        if not file_upload:
            file_upload = task.get('data', {}).get('ocr', '')

        if not file_upload:
            continue

        filename = file_upload.split('-')[-1]
        relative_path = f"train/{filename}"

        annotations = task.get('annotations', [])
        if not annotations:
            continue

        results = annotations[0].get('result', [])

        bboxes = {}
        transcriptions = {}

        # 각 박스별 실제 이미지 원본 크기(Resolution)를 담을 딕셔너리
        img_sizes = {}

        for res in results:
            res_id = res.get('id')
            res_type = res.get('type')
            value = res.get('value', {})

            # 버그 수정: original_width/height는 value 내부가 아니라 res 바로 아래에 있습니다!
            orig_w = res.get('original_width') or value.get('original_width') or 1
            orig_h = res.get('original_height') or value.get('original_height') or 1

            if res_type == 'rectangle':
                bboxes[res_id] = value
                img_sizes[res_id] = (orig_w, orig_h) # 해당 id의 이미지 크기 강제 매핑
            elif res_type == 'textarea':
                text_list = value.get('text', [])
                transcriptions[res_id] = text_list[0] if text_list else ""

        paddle_annotations = []

        for res_id, bbox in bboxes.items():
            text = transcriptions.get(res_id, "")

            # 저장해 둔 실제 해상도 크기를 가져옴
            orig_w, orig_h = img_sizes.get(res_id, (1, 1))

            # 이제 1을 곱하는 게 아니라 실제 이미지 해상도(예: 1498)를 정상적으로 곱합니다
            x = (bbox.get('x', 0) / 100.0) * orig_w
            y = (bbox.get('y', 0) / 100.0) * orig_h
            w = (bbox.get('width', 0) / 100.0) * orig_w
            h = (bbox.get('height', 0) / 100.0) * orig_h

            # 정상적인 시계방향 4개 점 좌표 생성
            points = [
                [round(x), round(y)],
                [round(x + w), round(y)],
                [round(x + w), round(y + h)],
                [round(x), round(y + h)]
            ]

            paddle_annotations.append({
                "transcription": text,
                "points": points
            })

        out_lines.append(f"{relative_path}\t{json.dumps(paddle_annotations, ensure_ascii=False)}\n")

    os.makedirs(os.path.dirname(OUTPUT_TXT_PATH), exist_ok=True)

    with open(OUTPUT_TXT_PATH, 'w', encoding='utf-8') as out_f:
        out_f.writelines(out_lines)
        out_f.flush()
        os.fsync(out_f.fileno())

    print("변환 프로세스 완료")
    print("생성된 라벨 파일 경로:", OUTPUT_TXT_PATH)
    print("총 처리된 이미지 수:", len(out_lines))
```

```python
import json
import os

BASE_PATH = '/content/drive/MyDrive/OCR'
TXT_PATH = os.path.join(BASE_PATH, 'train/train_txt.txt')

if not os.path.exists(TXT_PATH):
    print("오류: 검증할 train_txt.txt 파일이 존재하지 않습니다:", TXT_PATH)
else:
    with open(TXT_PATH, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    total_lines = len(lines)
    print(f"=== [검증 결과 요약] ===")
    print(f"총 추출된 이미지(줄) 수: {total_lines}장")

    if total_lines == 32:
        print("정상: 목표한 32장 데이터가 유실 없이 모두 포함되었습니다.")
    else:
        print(f"경고: 목표치(32장)와 일치하지 않습니다. (현재 {total_lines}장)")

    print("\n=== [상세 데이터 구조 검증] ===")
    error_count = 0

    for idx, line in enumerate(lines, 1):
        line = line.strip()
        if not line:
            print(f"줄 {idx}: 경고 - 빈 줄이 발견되었습니다.")
            continue

        # 탭 분리 검증
        parts = line.split('\t')
        if len(parts) != 2:
            print(f"줄 {idx}: 오류 - 이미지 경로와 라벨 데이터가 탭(\\t)으로 분리되지 않았습니다.")
            error_count += 1
            continue

        img_path, label_str = parts

        # JSON 구조 검증
        try:
            label_json = json.loads(label_str)
            box_count = len(label_json)

            # 상위 3개랑 하위 2개만 샘플로 로그 출력 (너무 길어지는 것 방지)
            if idx <= 3 or idx > total_lines - 2:
                print(f"  [{idx:02d}번 이미지] 경로: {img_path} | 추출된 글자 박스: {box_count}개")
            elif idx == 4:
                print("  ...")

        except json.JSONDecodeError:
            print(f"줄 {idx}: 오류 - 라벨 데이터가 올바른 JSON 형식이 아닙니다.")
            error_count += 1

    print("\n=== [최종 판단] ===")
    if error_count == 0:
        print("라벨 파일 검증 완료: 모든 데이터가 PaddleOCR 학습 규격에 100% 완벽하게 부합합니다! 다음 단계로 진행하셔도 좋습니다.")
    else:
        print(f"라벨 파일 검증 실패: 총 {error_count}개의 구조 오류가 발견되었습니다. 코드를 다시 확인해야 합니다.")
```

```python
import json
import os
import cv2

BASE_PATH = '/content/drive/MyDrive/OCR'
INPUT_JSON_PATH = os.path.join(BASE_PATH, 'train/label_studio_data.json')
ORIGINAL_IMG_DIR = os.path.join(BASE_PATH, 'train') # 원본 이미지 경로
RESIZE_IMG_DIR = os.path.join(BASE_PATH, 'train_resize') # 리사이즈된 이미지 저장 경로
OUTPUT_TXT_PATH = os.path.join(BASE_PATH, 'train/train_txt.txt')

# 설정: 변환할 이미지의 타겟 가로(Width) 크기 (메모리 방지를 위해 800~1000 추천)
TARGET_WIDTH = 800

os.makedirs(RESIZE_IMG_DIR, exist_ok=True)

with open(INPUT_JSON_PATH, 'r', encoding='utf-8') as f:
    data = json.load(f)

out_lines = []
resize_count = 0

for task in data:
    file_upload = task.get('file_upload', '')
    if not file_upload:
        file_upload = task.get('data', {}).get('ocr', '')
    if not file_upload:
        continue

    filename = file_upload.split('-')[-1]
    img_path = os.path.join(ORIGINAL_IMG_DIR, filename)

    if not os.path.exists(img_path):
        print(f"경고: 원본 이미지를 찾을 수 없습니다: {filename}")
        continue

    # 1. 이미지 로드 및 리사이즈 비율 계산
    img = cv2.imread(img_path)
    if img is None:
        continue
    orig_h, orig_w, _ = img.shape

    # 가로 크기 기준으로 세로 크기를 비율 맞게 계산 (Aspect Ratio 유지)
    ratio = TARGET_WIDTH / float(orig_w)
    target_height = int(orig_h * ratio)

    # 이미지 크기 축소 (버그 수정: 대문자 cv2.INTER_AREA 사용)
    resized_img = cv2.resize(img, (TARGET_WIDTH, target_height), interpolation=cv2.INTER_AREA)

    resize_filename = f"resize_{filename}"
    resize_save_path = os.path.join(RESIZE_IMG_DIR, resize_filename)
    cv2.imwrite(resize_save_path, resized_img)
    resize_count += 1

    # 2. 라벨 데이터 파싱 및 축소 비율 적용
    annotations = task.get('annotations', [])
    if not annotations:
        # 라벨이 없는 빈 이미지도 전체 구조 학습을 위해 포함
        relative_path = f"train_resize/{resize_filename}"
        out_lines.append(f"{relative_path}\t[]\n")
        continue

    results = annotations[0].get('result', [])

    bboxes = {}
    transcriptions = {}
    for res in results:
        res_id = res.get('id')
        res_type = res.get('type')
        value = res.get('value', {})

        # Label Studio 원본 등록 크기 체크
        task_w = res.get('original_width') or value.get('original_width') or orig_w
        task_h = res.get('original_height') or value.get('original_height') or orig_h

        if res_type == 'rectangle':
            bboxes[res_id] = (value, task_w, task_h)
        elif res_type == 'textarea':
            text_list = value.get('text', [])
            transcriptions[res_id] = text_list[0] if text_list else ""

    paddle_annotations = []

    # 3. 줄어든 이미지 크기에 맞춰 좌표 재계산
    for res_id, (bbox, tw, th) in bboxes.items():
        text = transcriptions.get(res_id, "")

        # 원본 해상도 기준 픽셀 좌표 선계산
        x_orig = (bbox.get('x', 0) / 100.0) * tw
        y_orig = (bbox.get('y', 0) / 100.0) * th
        w_orig = (bbox.get('width', 0) / 100.0) * tw
        h_orig = (bbox.get('height', 0) / 100.0) * th

        # 리사이즈 비율(ratio)을 곱해서 새 좌표 구하기
        new_x = x_orig * ratio
        new_y = y_orig * ratio
        new_w = w_orig * ratio
        new_h = h_orig * ratio

        # PaddleOCR용 시계방향 4개 점 생성
        points = [
            [round(new_x), round(new_y)],
            [round(new_x + new_w), round(new_y)],
            [round(new_x + new_w), round(new_y + new_h)],
            [round(new_x), round(new_y + new_h)]
        ]

        paddle_annotations.append({
            "transcription": text,
            "points": points
        })

    # 경로를 train_resize/... 로 변경하여 라벨 저장
    relative_path = f"train_resize/{resize_filename}"
    out_lines.append(f"{relative_path}\t{json.dumps(paddle_annotations, ensure_ascii=False)}\n")

# 최종 train_txt.txt 파일 업데이트
with open(OUTPUT_TXT_PATH, 'w', encoding='utf-8') as out_f:
    out_f.writelines(out_lines)
    out_f.flush()

print(f"=== 리사이즈 프로세스 완료 ===")
print(f"성공적으로 축소된 이미지 수: {resize_count}장")
print(f"새로운 이미지 해상도: {TARGET_WIDTH} x (비율 맞춤)")
print(f"리사이즈 이미지 저장 경로: {RESIZE_IMG_DIR}")
print(f"라벨 파일 업데이트 완료: {OUTPUT_TXT_PATH}")
```

```python
import os
import json

BASE_PATH = '/content/drive/MyDrive/OCR'
TXT_PATH = os.path.join(BASE_PATH, 'train/train_txt.txt')

print("=== [데이터셋 검증 시작] ===")

if not os.path.exists(TXT_PATH):
    print("오류: train_txt.txt 파일이 존재하지 않습니다.")
else:
    with open(TXT_PATH, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    print(f"총 라벨 수: {len(lines)} (목표: 32)")

    error_flag = False
    missing_count = 0
    coord_error_count = 0

    for idx, line in enumerate(lines, 1):
        line = line.strip()
        if not line:
            continue

        img_rel_path, label_str = line.split('\t')
        full_img_path = os.path.join(BASE_PATH, img_rel_path)

        # 1. 크롭/리사이즈 처리된 이미지 파일 존재 여부 검증
        if not os.path.exists(full_img_path):
            print(f"오류 - [{idx:02d}번] 이미지 파일 누락: {full_img_path}")
            missing_count += 1
            error_flag = True
            continue

        # 2. 변환 좌표의 유효 범위 검증 (해상도 800 기준)
        labels = json.loads(label_str)
        for label in labels:
            points = label.get('points', [])
            for pt in points:
                if pt[0] > 800 or pt[0] < 0:
                    print(f"오류 - [{idx:02d}번] 좌표가 유효 범위를 초과함: {pt}")
                    coord_error_count += 1
                    error_flag = True
                    break
            if error_flag:
                break

    print("\n=== [최종 검증 결과] ===")
    if not error_flag and len(lines) == 32:
        print("결과: 정상 (32장 이미지 및 라벨 좌표 일치 확인 완료)")
    else:
        print(f"결과: 실패 (누락 이미지: {missing_count}개, 좌표 오류: {coord_error_count}개)")
```
